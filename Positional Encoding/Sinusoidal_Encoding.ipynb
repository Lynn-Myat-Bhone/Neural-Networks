```python
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
def get_positional_encoding(seq_len,d_model):
    PE = np.zeros((seq_len,d_model))
    for pos in range(seq_len):
        for i in range(0,d_model,2): # step 2 to take 2 values from i , not for even and odd
            angle = pos/10000**(2*i/d_model)
            PE[pos,i]=(np.sin(angle) )  #even
            PE[pos,i+1]=(np.cos(angle)) #odd    
       
    return PE
```

```python
seq_len = 20
d_model = 16
res = get_positional_encoding(seq_len,d_model)
print(res.shape)
```

```python
plt.figure(figsize=(5,3))
plt.plot(res)
plt.show()
```
